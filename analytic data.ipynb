```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# in excel, =countblank(range)
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_tracks = df_tracks.sort_values(by='popularity', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='popularity', y='name', data=sorted_tracks, palette='viridis')
```

```python
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks[df_tracks['popularity'] > 90].sort_values(by='popularity', ascending=False)
most_popular[['name', 'popularity', 'artists']].iloc[:10]
```

```python
df_tracks.set_index('release_date', inplace=True)
```

```python
# df_tracks.index = pd.to_datetime(df_tracks.index, errors='coerce')     # convert data that contain only year into NaT
df_tracks.head(10)
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))  # Convert ms to s
df_tracks.drop(columns=['duration_ms'], inplace=True)
# df_tracks.drop(duration_ms, axis=1, inplace=True, errors='ignore')
```

```python
df_tracks['duration'].head()
```

```python
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis=1).select_dtypes(include='number').corr(method='pearson')
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title('Correlation Heatmap of Track Features')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004* len(df_tracks)), random_state=42)  # 0.4%, 4 out of every 1000 tracks
print(f'Sample size: {len(sample_df)}')
```

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='energy', y='loudness', data=sample_df, color="c").set_title('Energy vs Loudness')
```

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='acousticness', y='popularity', data=sample_df, color="b").set_title('Acousticness vs Popularity')
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'], errors='coerce')
years = df_tracks['dates'].dt.year
```

```python
sns.displot(years, discrete=True, kind="hist", color='purple', height=5, aspect=2).set(title='Song Distribution by Year')
```

```python
total_dr = df_tracks['duration']
sns.set_style("whitegrid")
plt.xticks(rotation=90)
fig_dims = (12, 6)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x=years.tolist(), y=total_dr.tolist(), ax=ax, color='orange').set(title='Total Duration of Songs by Year', xlabel='Year', ylabel='Total Duration (s)')
```
